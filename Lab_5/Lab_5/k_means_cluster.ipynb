```python
import matplotlib.pyplot as plt
import random

def k_means_clustering(data, k, iterations=10):
    centroids = data[:k]
    
    for _ in range(iterations):
        clusters = {i: [] for i in range(k)}
        
        for point in data:
            distances = [abs(point - centroid) for centroid in centroids]
            nearest_centroid_index = distances.index(min(distances))
            clusters[nearest_centroid_index].append(point)
        
        for idx in clusters:
            if clusters[idx]:
                centroids[idx] = sum(clusters[idx]) / len(clusters[idx])
    
    return clusters, centroids

def plot_clusters(clusters):
    colors = ['red', 'blue', 'green', 'purple', 'orange']
    
    for idx, points in clusters.items():
        y_values = [random.uniform(0, 1) * 50 for _ in points]  # Random Y values for style
        plt.scatter(points, y_values, color=colors[idx], label=f'Cluster {idx}')
    
    plt.xlabel('Data Points')
    plt.ylabel('Randomized Y for Display')
    plt.title('K-Means Clustering (Scatter Style)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = [2, 3, 4, 10, 11, 12, 20, 25, 30]
k = 2
clusters, centroids = k_means_clustering(data, k)

print("Clusters:", clusters)
print("Centroids:", centroids)

plot_clusters(clusters)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score


#given Actual and predicred Classes
actual_class = np.array([1, 0, 0, 1, 0, 1]) 
```

```python
actuall_class = np.array([1,0,0,1,0,1,0,0,1,0,1,0])
predicted_class = np.array([1,1,0,1,0,1,1,0,1,0,0,0])
```

```python
# step 2: Extract TP, TN, FP, FN
cm = confusion_matrix(actuall_class,predicted_class)
TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# step 3: Calculate performance matrix
accuracy_score = (TP+TN) / (TP + TN + FP + FN)
misclassification_rate = (FP + FN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) != 0 else 0
specificity = TN / (TN + FP) if (TN + FP)!= 0 else 0
```

```python
# step 4: Display the Result
print("Confusion Matrix:")
print(cm)
print("\n Performance Metrices:")
print(f"Accuracy: {accuracy_score:.2f}")
print(f"Misclassification Rate: {misclassification_rate:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Specificity: {specificity:.2f}")
```

```python
# step 5: Visualize the confusion matrix

def plot_confusion_matrix(cm, title='Confusion Matrix'):
    plt.figure(figsize=(8,6)) 
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Predicted: 0', 'Predicted: 1'],
                yticklabels=['Actual: 0', 'Actual: 1'])
    
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
# Plot the confusion matrices
plot_confusion_matrix(cm, title='Confusion Matrix for Binary  Classification')
```
